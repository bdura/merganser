```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import bezier
```

```python
controls = np.array([
    [0, 1],
    [1, 0],
    [2, 3],
    [3, 0],
])
```

```python
ts = np.linspace(0, 1, 100)
num_curves = 100
curves = np.zeros((num_curves, ts.size, 2))

for i in range(num_curves):
    random_controls = controls + 0.1 * np.random.randn(*controls.shape)
    curves[i] = np.asarray([bezier.compute(t, random_controls) for t in ts])
```

```python
plt.figure(figsize=(10, 5))
ax = plt.subplot(1, 1, 1)

ax.scatter(*controls.T, c='C3')
for curve in curves:
    ax.plot(*curve.T, c='C0', alpha=0.05)
plt.show()
```

```python
ts = np.linspace(0, 1, 100)
num_curves = 100
curves = np.zeros((num_curves, ts.size, 2))

plt.figure(figsize=(7, 7))
ax = plt.subplot(1, 1, 1)

for _ in range(num_curves):
    
    random_controls = controls + 0.1 * np.random.randn(*controls.shape)
    curve = np.asarray([bezier.compute(t, random_controls) for t in ts])
    
    ax.scatter(*random_controls.T, c='C3', alpha=.1)
    ax.plot(*curve.T, c='C0', alpha=0.05)
    
plt.axis('scaled')
plt.show()
```
